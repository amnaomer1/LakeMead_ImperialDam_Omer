## **1. Introduction**

The Colorado River is one of the most heavily managed and over-allocated river systems in the world, sustaining more than 40 million people and irrigating over two million hectares (≈5 million acres) of cropland across the southwestern United States and northern Mexico (Richter et al., 2024). Persistent overuse and long-term drought have caused water consumption to exceed natural runoff in 16 of 21 years between 2000 and 2020, leading to major declines in storage across the basin’s two largest reservoirs—Lake Powell and Lake Mead (Schmidt et al., 2023; Richter et al., 2024). This imbalance has forced policymakers to implement a series of shortage agreements and emergency conservation measures, including the 2007 Interim Guidelines,2019 Drought Contingency Plan, and 2023 System Conservation Proposal, to prevent system collapse (U.S. Bureau of Reclamation, 2019, 2023; U.S. Department of the Interior, 2023).

California is the largest single user of Colorado River water, accounting for the majority of consumptive use in the Lower Basin (Richter et al., 2024; U.S. Bureau of Reclamation, 2023). Consumptive use refers to water that is permanently removed from the river system and cannot be returned, primarily through crop evapotranspiration, evaporation, and other losses. Although cities also divert Colorado River water, most municipal indoor use is treated and returned to the system, meaning that agriculture represents the dominant share of consumptive use in the basin.

In California, most Colorado River water supports irrigated agriculture in the Imperial and Coachella Valleys, where crops such as alfalfa and other grass hays—used primarily as cattle feed—consume more than half of all water diverted from the river. Irrigated agriculture accounts for approximately 74% of direct human water use and 52% of total basin-wide consumption, underscoring California’s heavy dependence on consistent reservoir releases (Richter et al., 2024). Consequently, even small fluctuations in Lake Mead’s elevation can significantly impact irrigation reliability, agricultural productivity, and water management planning.

Lake Mead, the largest reservoir in the United States, functions as the Lower Basin’s principal storage and regulatory hub (U.S. Bureau of Reclamation, 2025). Its elevation defines federally mandated shortage tiers under the 2007 Interim Guidelines and the 2019 Drought Contingency Plan, determining how much water each state receives (U.S. Bureau of Reclamation, 2019, 2025). As Lake Mead levels decline, mandated cutbacks intensify, reducing the volume of water delivered to California through the All-American Canal, the primary diversion structure conveying Colorado River water to California’s agricultural districts (Stene, 2009; U.S. Bureau of Reclamation, 2023).

Understanding how changes in Lake Mead’s elevation affect California’s downstream diversions and consumptive use is vital for linking hydrologic variability with institutional management in the Lower Colorado River Basin. Fluctuations in reservoir elevation, driven by climate and inflow variability, translate into changes in downstream water allocation and reliability (Shin et al., 2019; Li et al., 2024). Quantifying the timing and strength of these linkages helps clarify how physical water-supply signals propagate through operational and policy frameworks that govern water availability (Garcia et al., 2020).

From a management perspective, this understanding provides an empirical basis for adaptive planning as shortage tiers increasingly determine how limited water is distributed among states and users (Wang & Rosenberg, 2023). For hydrologists and policymakers, it offers insight into how reservoir behavior and institutional rules jointly shape water outcomes. For agricultural communities, it highlights the growing challenge of maintaining productivity under declining storage and rising uncertainty.  

Overall, this project contributes to broader efforts to link reservoir dynamics, shortage policies, and real water-use outcomes, supporting resilience-oriented water governance in the Lower Colorado River Basin (Garza-Diaz & Sandoval-Solis, 2022; Hite et al., 2025; Richter et al., 2024).

To investigate these linkages, this study analyzes historical hydrologic and water-use records from 2000–2024, applying correlation and lag analysis, response elasticity evaluation, and policy-based before–after comparisons to assess how changes in Lake Mead’s elevation influence California’s diversions and consumptive use.

### **Research Question**  
How do changes in **Lake Mead’s elevation** between 2000 and 2024 influence **California’s annual water diversions and consumptive use**, as reported by the U.S. Bureau of Reclamation?

### **Hypothesis**  
Declines in Lake Mead’s elevation will result in reduced water deliveries and lower consumptive use in California, with time-lagged responses as operational and policy decisions adjust. When reservoir levels are high, the relationship between elevation and water use is primarily hydrologic and relatively stable. As elevation declines into federally defined shortage tiers, policy constraints increasingly regulate deliveries, reducing variability and slowing recovery. This pattern reflects the combined influence of hydrologic variability and institutional management in shaping how scarcity signals propagate through the Lower Colorado River system.

## **2. Study Site Description**

This project focuses on the **Lower Colorado River Basin**, specifically the hydrologic and operational link between **Lake Mead**—the system’s main storage reservoir—and **Imperial Dam**, the primary diversion point for California. Together, these two locations define the upstream–downstream boundary of the study, representing the physical and institutional connection through which Colorado River water moves from federal storage to state-level use (U.S. Bureau of Reclamation, 2023; Congressional Research Service, 2023). 

### **Spatial Boundaries**

- **Upstream boundary:** *Lake Mead* at Hoover Dam (Arizona–Nevada border), the main regulating reservoir for the Lower Basin. Lake Mead was formed by Hoover Dam, completed in 1936 (U.S. Bureau of Reclamation, 2025).

- **Downstream boundary:** *Imperial Dam*, located 303 miles (488 km) downstream of Hoover Dam near Yuma, Arizona. At this point, Colorado River water is diverted westward into the All-American Canal. Imperial Dam was completed in 1938 (Stene, 2009).

- **Receiving region and distribution system:**  
  - The **All-American Canal** (≈80 miles) conveys water from Imperial Dam to agricultural users in the **Imperial Valley**, delivering about **3.0 million acre-feet per year**, or roughly **three-quarters** of California’s Colorado River diversions.  
  - The **Coachella Canal**, a branch of the All-American Canal, supplies about **0.3 million acre-feet/year** to the **Coachella Valley Water District (CVWD)**.  
  - The **Colorado River Aqueduct**, operated by the **Metropolitan Water District of Southern California (MWD)**, conveys about **1.0 million acre-feet/year** to urban areas across southern California.  
  - Collectively, these conveyance systems comprise California’s ≈ **4.4 million acre-feet** Lower Basin allocation (U.S. Bureau of Reclamation, 2024; IID, 2025).


This corridor links federal reservoir operations with California’s consumptive use, forming the analytical focus for this study. This upstream–downstream linkage provides a natural framework for evaluating how changes in Lake Mead’s elevation—the indicator of hydrologic and policy conditions—propagate to California’s water diversions and consumptive use (U.S. Bureau of Reclamation, 2024). Because California is the largest single user of Colorado River water, accounting for most of the Lower Basin’s consumptive use (Richter et al., 2024), this setting offers both representativeness and sensitivity: it reflects basin-wide management dynamics while also revealing the local consequences of shortage-tier operations on agricultural and municipal water-supply reliability (Hite et al., 2025; U.S. Bureau of Reclamation, 2024).

### **4.1 Correlation & Lag Analysis**

#### 4.1.1 Method Overview

This analysis evaluates how fluctuations in Lake Mead’s elevation relate to California’s total consumptive water use (CU) from 2000–2024, and whether changes in upstream reservoir storage are followed by delayed downstream adjustments. Monthly Lake Mead elevation and CU datasets were merged for the common period and synchronized using month-end timestamps, retaining only months with valid values for both variables.

To isolate interannual variability, the mean seasonal cycle was removed from each series and the resulting anomalies were standardized into z-scores:

$$
z = \frac{x - \bar{x}}{s}
$$

where $x$ is a monthly value, $\bar{x}$ is the sample mean, and $s$ is the sample standard deviation. 
This approach expresses both variables as standardized deviations from their normal seasonal behavior, allowing direct, unitless comparison of how the two fluctuate through time.

This transformation enables direct comparison between elevation (ft) and consumptive use (acre-ft) on a unit-less scale.

Lagged Pearson correlations were then computed across a ±12-month window to assess response timing, with Spearman correlations and partial correlations (controlling for linear trend) used as robustness checks. Statistical significance was evaluated using Holm–Bonferroni correction to account for multiple lag tests.

Two diagnostic figures are produced:
1. A time-series plot of Lake Mead and CU seasonal-anomaly z-scores to visualize co-fluctuation patterns. 
2. A lag–correlation plot showing Pearson \( r \) versus lag with 95% confidence intervals and markers for statistically significant lags.

Together, these results characterize the timing and strength of the upstream–downstream connection between Lake Mead storage dynamics and California’s consumptive use.


#### 4.1.2 Running the Lag–Correlation Analysis

The following cell runs the core lag–correlation workflow using functions from the `lakemead_functions` module. The analysis aligns the two monthly datasets, filters to the defined study period, removes the mean seasonal cycle, and converts Lake Mead elevation and California consumptive use to standardized anomaly **z-scores**. It then computes lagged Pearson and Spearman correlations across a ±12-month window, including a Holm–Bonferroni adjustment to identify statistically significant relationships.

Two diagnostic plots are produced:

1. **Standardized Z-score Time Series** — illustrates co-fluctuation patterns between Lake Mead elevation and California consumptive use.
2. **Lag–Correlation Curve** — shows how correlation strength changes across time offsets and identifies the lag at which upstream reservoir conditions best relate to downstream water-use response.

Run the cell below to execute the analysis and generate both figures:




```python
# Run core lag analysis and plotting functions


df_clean, lagdf = lm. compute_lag_correlation_mead_cu(
    mead,
    water_raw,
    max_lag=12,
    study_start="2000-01-01",
    study_end="2024-12-31",
)


lm. plot_mead_cu_zscores(df_clean)
lm. plot_lag_correlation(lagdf)
```

This lag–correlation analysis evaluates how monthly fluctuations in Lake Mead’s elevation relate to California’s consumptive water use (CU) from 2000–2024, after removing recurring seasonal cycles. Correlations were examined across a ±12-month lag window to understand whether upstream reservoir storage conditions lead, follow, or co-vary with downstream water use decisions.

- The effective sample size is approximately 60, supporting statistical reliability despite autocorrelation in monthly time series.

The correlation curve shows a gradual strengthening moving from negative to positive lags, forming a broad and stable peak rather than a sharp spike. This suggests:

Thus, the interpretation is not that only a single lag matters, but rather that there is a persistent and sustained upstream-to-downstream connection across multiple operational time scales.


**Overall Interpretation**

The lag–correlation results reveal a strong and statistically supported linkage between Lake Mead storage levels and California’s consumptive water use. Changes in reservoir elevation appear to act as an early indicator of downstream behavioral and policy responses, with the most pronounced effect occurring about one year later, but with meaningful influence beginning almost immediately.

In practical terms, California’s water-use decisions respond to reservoir conditions with both short-term sensitivity and annual-scale adjustment behavior, consistent with agricultural planning cycles, irrigation management, and Colorado River allocation processes.


**Rationale for Monthly vs Annual Analysis and Seasonal Standardization**

Although the California consumptive-use dataset was compiled manually from annual USBR water-use reports, the underlying data represent monthly accounting volumes, and therefore can be analyzed at a monthly resolution. Using monthly values is appropriate for this study because reservoir elevation and agricultural water deliveries are both managed on a monthly operational cycle, and monthly variability provides meaningful information about short-term hydrologic response and irrigation adjustments.

Seasonal standardization (monthly z-scores) was applied to remove strong recurring seasonal patterns, which otherwise inflate correlations simply due to shared annual cycles rather than true interannual hydrologic relationships. Standardizing by month enables clearer interpretation of how anomalies deviate from typical seasonal behavior and allows comparison across variables with different units (feet versus acre-feet).

To evaluate whether this preprocessing altered the correlation structure, a sensitivity test was conducted by repeating the lagged correlation analysis using raw monthly values without seasonal adjustment. The timing of peak correlation remained consistent at approximately a 10–12 month lag, demonstrating that the observed upstream–downstream response is robust and not dependent on the seasonal adjustment method.

### **4.2 Response Elasticity Analysis** 

This analysis examines the elasticity of California’s consumptive water use in response to changes in Lake Mead elevation, using standardized anomalies to assess how strongly and proportionally water-use behavior adjusts to hydrologic variability.  
Two timescales were evaluated: monthly (to capture short-term, operational adjustments) and annual(to reveal longer-term, system-wide responses).

#### 4.2.1 Method Overview

Response elasticity is represented by the slope of a linear regression between standardized anomalies of Lake Mead elevation and consumptive use:

$$
z_{\text{CU}} = a + b \, z_{\text{Mead}}
$$

where:

- $z_{\text{CU}}$ = standardized anomaly in California’s consumptive use  
- $z_{\text{Mead}}$ = standardized anomaly in Lake Mead elevation  
- $b$ = elasticity coefficient (slope), representing the magnitude of change in water use per unit change in elevation anomaly  

A slope ($b$) near 1 indicates a proportional response (one-for-one change), while smaller slopes indicate a dampened or buffered response.

Monthly elasticity was calculated using seasonal z-scores. Each month’s data was standardized relative to its own long-term monthly mean and variability (e.g., all Januaries, all Februaries). This removes recurring seasonal effects and isolates how month-to-month deviations in reservoir elevation relate to deviations in water use under typical operating conditions. A simple linear regression between the monthly standardized series estimates the short-term elasticity of response.

Annual elasticity was estimated by averaging Lake Mead elevation and total consumptive use by calendar year and then standardizing across years using overall z-scores. This regression represents interannual elasticity, capturing year-to-year adjustments influenced by longer-term hydrologic and policy conditions.

Comparing monthly and annual elasticity coefficients clarifies response timing: monthly elasticity reflects near-term operational adjustments within irrigation seasons, while annual elasticity captures slower, cumulative adjustments linked to hydrologic conditions and policy interventions.


#### 4.2.2 Running the Analysis

The code below calls a custom functions from the `lakemead_functions` that computes elasticities on both monthly and annual timescales and generates regression plots comparing standardized Lake Mead elevation and consumptive use:


```python
# ---  Response Elasticity Analysis ---

df_monthly, df_annual, stats_elast = lm.compute_response_elasticity(
    mead,
    water_raw,
    study_start="2000-01-01",
    study_end="2024-12-31",
)

lm.plot_response_elasticity(df_monthly, df_annual, stats_elast)
plt.show()
 
```



#### 4.2.3 Results – Response Elasticity Analysis

The response elasticity analysis quantifies how changes in Lake Mead elevation align with changes in California’s consumptive water use when both are expressed as standardized anomalies (z‐scores). The regression slope (b) represents the proportional change in consumptive use relative to a one standard‐deviation shift in reservoir elevation.

**Monthly Elasticity**
The monthly regression (b = 0.48, R² = 0.23) shows a moderate positive relationship: months with higher‐than‐average Lake Mead elevation tend to coincide with slightly higher consumptive use. The relatively low explanatory power (R²) indicates that month‐to‐month variability is strongly influenced by short‐term operational factors such as crop irrigation schedules, precipitation events, delivery timing, and system buffering rather than immediate changes in reservoir storage. This suggests that monthly consumptive use does not respond directly or rapidly to short‐term fluctuations in Lake Mead elevation.

**Annual Elasticity**
When data are aggregated to yearly totals, the annual regression (b = 0.73, R² = 0.53) reveals a substantially stronger relationship. Years with lower‐than‐normal reservoir elevations correspond to noticeably reduced consumptive use, consistent with allocation adjustments, conservation measures, and longer‐term management decisions. The higher R² value indicates that the upstream–downstream connection becomes more apparent once short‐term noise is smoothed out.

**Summary Interpretation**
Together, these results indicate that California’s consumptive water use is more responsive to Lake Mead storage at annual rather than monthly timescales. Short‐term operational flexibility buffers monthly fluctuations, while longer‐term planning, water‐ordering decisions, and policy responses translate reservoir conditions into observable reductions in use over the course of a year. This pattern aligns with how irrigation districts and agricultural users along the All‐American Canal structure seasonal and annual water‐management decisions.

### **4.3 Policy Milestones and Before–After Water Use Analysis**

This analysis evaluates how California’s total Colorado River **diversions** and **consumptive use** changed between 2000 and 2024 in relation to major federal and basin-level water management policies.  
The objective is to assess whether significant policy interventions correspond to measurable shifts in annual water-use behavior and system-wide conservation outcomes.


#### 4.3.1 Method Description

Monthly diversion and consumptive-use values were aggregated into annual totals, retaining only complete calendar years (12 months of data). Annual values were converted from acre-feet (af) to **million acre-feet per year (MAF/year)** to align with standard federal accounting metrics used in basin planning.

A set of policy milestones was selected to evaluate before–after changes in California’s water-use patterns:

- **Quantification Settlement Agreement (QSA, 2003)**  
  Established California’s legal entitlement to **4.4 MAF/year** from the Colorado River, formalized major intra-state water transfers, and funded efficiency improvements such as canal lining (State of California et al., 2003).

- **Interim Guidelines (2007)**  
  Defined shortage-sharing rules and coordinated operations for Lake Powell and Lake Mead. Introduced **Intentionally Created Surplus (ICS)** storage mechanisms (U.S. Bureau of Reclamation, 2007).

- **System Conservation Pilot Program (2014–2017)**  
  A voluntary, incentive-based program compensating water users—including California irrigation districts—for conserving water that remained stored in Lake Mead (U.S. Bureau of Reclamation, 2014).

- **Minute 323 (2017)**  
  A binational U.S.–Mexico agreement expanding ICS eligibility and financing conservation and fallowing projects that directly supported Lake Mead elevations (Hite et al., 2025).

- **Drought Contingency Plan (DCP, 2019)**  
  Strengthened shortage contributions and revised ICS accounting. Marked California’s first formal commitment to reductions under specified hydrologic conditions (Congressional Research Service, 2019).

- **Federal Shortage Declaration (2021)**  
  Declared due to critically low reservoir elevations, heightening drought response and accelerating voluntary conservation and ICS banking strategies (U.S. Department of the Interior, 2022a).

- **Tier 1 Shortage Implementation (2022)**  
  Reinforced mandatory reductions for other states and intensified conservation expectations across the Lower Basin.

- **Lower Colorado River Basin System Conservation and Efficiency Program (2022–present)**  
  A new federal incentive program compensating water users, including California agricultural districts, for voluntary multi-year water savings and efficiency actions aimed at stabilizing Lake Mead (U.S. Department of the Interior, 2022b).

Each milestone year is represented in the time-series plot with vertical markers, with extended program periods shown as shaded bands.


#### Trend Visualization

Annual time-series plots of **diversions** and **consumptive use** quantify how both values evolved under successive policy regimes.  
Using a shared MAF/year axis enables direct comparison of total withdrawals versus actual depletion, revealing patterns in efficiency, storage behavior, and responses to regulatory pressures.

These results illustrate the critical role of coordinated governance and economic instruments in balancing water security, agricultural production, and long-term system sustainability.

## **5.  Conclusion and Implications**

This study examined how changes in **Lake Mead’s elevation** between 2000 and 2024 influenced **California’s annual water diversions and consumptive use**, revealing a strong and measurable upstream–downstream connection within the Lower Colorado River system. Across multiple analytical approaches—policy evaluation, elasticity estimation, and lag-correlation analysis—the results consistently support the hypothesis that declining reservoir levels drive reductions in California’s water use through both hydrologic scarcity signals and institutional policy responses.

The correlation analysis demonstrates a strong and statistically significant upstream–downstream linkage between Lake Mead elevation and California’s consumptive use. The lag–correlation curve shows consistently high positive correlations across a wide band of positive lags (approximately +1 to +12 months), with values ranging from r ≈ 0.45–0.52 and peaking around +10–12 months. This indicates that changes in Lake Mead elevation influence downstream water‐use behavior across multiple operational timeframes—not only one distinct delay—and that reservoir conditions function as an early signal rather than triggering an immediate response.

These temporal dynamics align with the structure of decision processes in the Lower Colorado River system: agricultural water ordering cycles, annual allocation planning, and shortage declarations occur seasonally and yearly, producing gradual and sustained adjustment rather than rapid month‐to‐month reactions.

The elasticity results reinforce this interpretation. Monthly elasticity estimates show only moderate explanatory power (b ≈ 0.48, R² = 0.23), reflecting short‐term buffering by irrigation scheduling, weather variability, and operational flexibility. However, the annual elasticity relationship is substantially stronger (b ≈ 0.73, R² = 0.53), demonstrating that long-term storage deficits translate into reduced consumptive use more clearly once short-term noise is averaged out.

Finally, the policy impact results indicate that institutional actions amplify hydrologic scarcity signals. ΔMean (%) reductions in water use were largest after major management interventions, particularly the QSA (2003) and Post-2022 Incentive Programs, with meaningful declines also following the 2007 Interim Guidelines and the 2014 System Conservation Pilot Program.

These findings underscore that declining Lake Mead levels did not act in isolation; rather, coordinated policy responses played a critical role in driving and sustaining reductions in CU and Diversions under worsening drought conditions.


### Implications

These findings have several important implications:

- **For water-resource scientists**  
  The presence of strong and persistent correlations across +1 to +12 months suggests that modeling frameworks should incorporate multi-month and annual-scale response dynamics, linking storage conditions to human decision processes. Predictive tools may benefit from integrating reservoir outlooks, policy triggers, and crop cycles rather than relying solely on short-term climate or demand indicators.

- **For watershed and irrigation managers**  
  The temporal structure of the relationship provides a meaningful planning horizon: reservoir trends offer advance signals that can be used to adjust water-ordering decisions, conservation measures, and crop selection before shortages fully materialize. This supports proactive, rather than reactive, operational management.

- **For policy makers and basin negotiators**  
  The ability of consumptive use to adjust substantially over longer time scales reinforces the effectiveness of long-lead allocation policies, shortage tiers, and incentive-based conservation programs. Future agreements may benefit from prioritizing flexibility, financial mechanisms, and adaptive schedules aligned with evolving storage conditions.

- **For agricultural and community stakeholders**  
  Strategic reservoir-informed planning can help maintain food production and economic stability even under hydrologic stress, reducing conflict and uncertainty. However, this requires continued commitment to long-term conservation strategies rather than short-term crisis responses.


### **Speculation**
If reservoir declines accelerate beyond current projections, the effectiveness of buffering mechanisms may diminish, and more aggressive conservation or demand-reduction strategies may be needed. Conversely, climate-driven improvements in inflows combined with sustained conservation could support gradual storage recovery.


**In summary**, the results confirm that California’s water-use behavior is tightly linked to Lake Mead storage trends and increasingly shaped by governance frameworks. Understanding the timing and magnitude of this response is essential for designing future drought-resilience strategies in the Colorado River Basin.
