## 1. Introduction

The Colorado River is one of the most heavily managed and over-allocated river systems in the world, sustaining more than 40 million people and irrigating over two million hectares (≈5 million acres) of cropland across the southwestern United States and northern Mexico (Richter et al., 2024). Persistent overuse and long-term drought have caused water consumption to exceed natural runoff in 16 of 21 years between 2000 and 2020, leading to major declines in storage across the basin’s two largest reservoirs—Lake Powell and Lake Mead (Schmidt et al., 2023; Richter et al., 2024). This imbalance has forced policymakers to implement a series of shortage agreements and emergency conservation measures, including the 2007 Interim Guidelines,2019 Drought Contingency Plan, and 2023 System Conservation Proposal, to prevent system collapse (U.S. Bureau of Reclamation, 2019, 2023; U.S. Department of the Interior, 2023).

Within this system, California is the largest single user of Colorado River water, accounting for the majority of consumptive use in the Lower Basin (Richter et al., 2024; U.S. Bureau of Reclamation, 2023). Most of this water supports irrigated agriculture in the Imperial and Coachella Valleys, where crops such as alfalfa and other grass hays—used primarily as cattle feed—consume more than half of all water diverted from the river. Irrigated agriculture accounts for roughly 74% of direct human water use and 52% of total basin-wide consumption, emphasizing the scale of California’s dependence on consistent reservoir releases (Richter et al., 2024). As a result, even small fluctuations in Lake Mead’s elevation have direct consequences for the state’s irrigation reliability, agricultural productivity, and water management planning.

Lake Mead, the largest reservoir in the United States, functions as the Lower Basin’s principal storage and regulatory hub (U.S. Bureau of Reclamation, 2025). Its elevation defines federally mandated shortage tiers under the 2007 Interim Guidelines and the 2019 Drought Contingency Plan, determining how much water each state receives (U.S. Bureau of Reclamation, 2019, 2025). As Lake Mead levels decline, mandated cutbacks intensify, reducing the volume of water delivered to California through the All-American Canal—the primary diversion structure conveying Colorado River water to California’s agricultural districts (Stene, 2009; U.S. Bureau of Reclamation, 2023).


Understanding how changes in Lake Mead’s elevation affect California’s downstream diversions and consumptive use is vital for linking hydrologic variability with institutional management in the Lower Colorado River Basin. Fluctuations in reservoir elevation—driven by climate and inflow variability—translate into changes in downstream water allocation and reliability (Shin et al., 2019; Li et al., 2024). Quantifying the timing and strength of these linkages helps clarify how physical water-supply signals propagate through operational and policy frameworks that govern water availability (Garcia et al., 2020).

From a management perspective, this understanding provides an empirical basis for adaptive planning as shortage tiers increasingly determine how limited water is distributed among states and users (Wang & Rosenberg, 2023). For hydrologists and policymakers, it offers insight into how reservoir behavior and institutional rules jointly shape water outcomes. For agricultural communities, it highlights the growing challenge of maintaining productivity under declining storage and rising uncertainty.  

Overall, this project contributes to broader efforts to link reservoir dynamics, shortage policies, and real water-use outcomes, supporting resilience-oriented water governance in the Lower Colorado River Basin (Garza-Diaz & Sandoval-Solis, 2022; Congressional Research Service, 2023; Richter et al., 2024).



**Research Question**  
How do changes in **Lake Mead’s elevation** between 2000 and 2024 influence **California’s annual water diversions and consumptive use**, as reported by the U.S. Bureau of Reclamation?

**Hypothesis**  
Fluctuations in Lake Mead’s elevation generate measurable, time-lagged responses in California’s downstream water deliveries and consumptive use.  
When reservoir levels are high, the relationship between elevation and use is primarily hydrologic and relatively stable. As Lake Mead’s elevation declines and shortage tiers are triggered, policy constraints increasingly regulate deliveries, reducing variability and slowing recovery.  
This pattern reflects the combined influence of hydrologic variability and institutional management in shaping how scarcity signals propagate through the Lower Colorado River system.

### Loading and Visualizing Raw Data

This section loads the two primary datasets, converts the date columns to datetime format, and generates preliminary time-series plots to visualize variability and temporal coverage.


**Lake Mead Elevation: Annual End-of-Year Time Series**

This section extracts and visualizes **end-of-year elevations** for Lake Mead from 2000 to 2024, based on monthly records reported by the U.S. Bureau of Reclamation for **Lake Mead at Hoover Dam**.  
Since the dataset already contains **end-of-month values**, selecting the December observation from each year provides a direct representation of the **annual closing elevation**—a standard indicator of reservoir storage status used in Reclamation reports.

The following steps are performed in the code below:
1. Load the monthly elevation dataset and ensure the columns are correctly formatted.  
2. Filter the data to include only December records (end of each calendar year).  
3. Plot the resulting **annual time series** to visualize long-term hydrologic variability, highlighting years of decline and partial recovery.  

This annual plot simplifies interpretation of multi-year storage trends and supports later analyses linking Lake Mead’s hydrologic state to California’s diversions and consumptive use.


```python
# --- Lake Mead Elevation (End-of-Year Plot) ---

import pandas as pd
import matplotlib.pyplot as plt

# File path
mead_path = "data/LakeMead_Elevation_Monthly_2000_2024.csv"

# Load and prepare data
mead = pd.read_csv(mead_path, parse_dates=["Date"])
mead = mead.sort_values("Date").reset_index(drop=True)

# Extract December (end-of-year) values
mead["Month"] = mead["Date"].dt.month
annual_end = mead[mead["Month"] == 12].copy()
annual_end["Year"] = annual_end["Date"].dt.year

# Quick check
display(annual_end[["Year", "Elevation_ft"]].head())

# Plot: end-of-year elevation
plt.figure(figsize=(10,5))
plt.plot(annual_end["Year"], annual_end["Elevation_ft"], lw=2, marker="o")
plt.title("Lake Mead Elevation (End-of-Year, 2000–2024)")
plt.xlabel("Year")
plt.ylabel("Elevation (ft)")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

**California Water Use: Data Loading, Cleaning, and Annual Aggregation**

This section compiles and visualizes California’s total Colorado River water use between 2000 and 2024, using data manually extracted from the *U.S. Bureau of Reclamation’s Colorado River Accounting and Water Use Reports*.  
The dataset includes monthly values for **diversions**, **measured returns**, **unmeasured returns**, and **total consumptive use**, all reported in **acre-feet**.

The following steps are performed in the code below:
1. Load the CSV file containing monthly water-use data (2000–2024).  
2. Clean numeric fields by removing formatting characters (commas) and converting them to numeric data types.  
3. Identify complete calendar years (12 months present) to ensure consistency across the time series.  
4. Aggregate the monthly data to **annual totals** for both **diversion** and **consumptive use**.  
5. Plot the results to illustrate interannual variability in California’s water diversions and use.

Because the Bureau of Reclamation reports water-use data on a **calendar-year basis**, the analysis also uses calendar years.  
This ensures alignment with both the Lake Mead elevation data and Reclamation’s official accounting framework, allowing consistent interpretation of upstream–downstream relationships.


```python
# ---  California Water Use (Load + Plot) ---

import pandas as pd
import matplotlib.pyplot as plt

# File path
wateruse_path = "data/California_WaterUse_2000_2024.csv"

# Load as strings to safely strip thousands separators
wateruse_raw = pd.read_csv(wateruse_path, dtype=str)

# Keep expected columns 
expected_cols = ["Date", "Diversion_af", "MeasuredReturns_af", "UnmeasuredReturns_af", "ConsumptiveUse_af"]
present = [c for c in expected_cols if c in wateruse_raw.columns]
if not {"Date", "Diversion_af", "ConsumptiveUse_af"}.issubset(present):
    raise ValueError("Water-use file must include at least: Date, Diversion_af, ConsumptiveUse_af")

wateruse = wateruse_raw[present].copy()

# Parse dates
wateruse["Date"] = pd.to_datetime(wateruse["Date"], errors="coerce")
wateruse = wateruse.dropna(subset=["Date"]).sort_values("Date").reset_index(drop=True)

# Convert numeric columns (remove commas, coerce to numbers)
num_cols = [c for c in wateruse.columns if c.endswith("_af")]
for c in num_cols:
    wateruse[c] = (
        wateruse[c]
        .astype(str)
        .str.replace(",", "", regex=False)
        .str.strip()
        .replace({"": None})
    )
    wateruse[c] = pd.to_numeric(wateruse[c], errors="coerce")

# Quick check
display(wateruse.head())


# ---------- Annual totals (only complete years with 12 months) ----------
wateruse["Year"] = wateruse["Date"].dt.year
months_per_year = wateruse.groupby("Year")["Date"].nunique()
complete_years = months_per_year[months_per_year == 12].index

annual = (
    wateruse.loc[wateruse["Year"].isin(complete_years)]
    .groupby("Year", as_index=False)[["Diversion_af", "ConsumptiveUse_af"]]
    .sum(min_count=1)
    .sort_values("Year")
)

display(annual.head())

plt.figure(figsize=(10,6))
plt.plot(annual["Year"], annual["Diversion_af"], label="Diversion (af/yr)", lw=2)
if "ConsumptiveUse_af" in annual.columns:
    plt.plot(annual["Year"], annual["ConsumptiveUse_af"], label="Consumptive Use (af/yr)", lw=2)
plt.title("California Water Use (Annual Totals, complete years only)")
plt.xlabel("Year")
plt.ylabel("Acre-Feet per Year")
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

```

### 4.1 Correlation & Lag Analysis  

This analysis examines how fluctuations in **Lake Mead’s elevation** relate to **California’s total consumptive water use (CU)** from 2000 to 2024.  
The goal is to identify whether changes in upstream reservoir storage conditions are followed by delayed adjustments in downstream water-use behavior and to quantify the timing of this relationship.

**Data Preparation and Alignment**  
Monthly datasets for Lake Mead elevation and California’s consumptive use were combined for the same study period.  
To ensure perfect temporal consistency, both datasets were converted to month-end timestamps and merged by date so that each record represents a synchronized pair of monthly values.  
Only months containing valid data for both variables were retained.

A water year (WY) column was also created (October–September) to facilitate later hydrologic interpretation and allow results to be grouped by operational years if needed.

**Seasonal Adjustment and Standardization**  
Because both variables have strong seasonal cycles, the analysis first removed the monthly climatology to isolate interannual variability.  
For each variable, monthly anomalies were computed by subtracting the long-term mean for that month, followed by standardization into z-scores:

Because Lake Mead elevation (measured in feet) and consumptive use (measured in acre-feet) have different units and scales, both variables were standardized using z-scores:

$$
z = \frac{x - \bar{x}}{s}
$$

where $x$ is a monthly value, $\bar{x}$ is the sample mean, and $s$ is the sample standard deviation. 
This approach expresses both variables as standardized deviations from their normal seasonal behavior, allowing direct, unitless comparison of how the two fluctuate through time.

**Correlation Estimation**  
The statistical relationship between the standardized series was measured using the Pearson correlation coefficient (r), which quantifies the strength and direction of their linear association.  
To ensure reliability under temporal dependence, the analysis used an effective sample size (nₑ) that adjusts for autocorrelation in both time series.  
For each lag, the following metrics were computed:
- Pearson correlation coefficient (r)  
- p-value testing the null hypothesis of no correlation  
- 95% confidence interval (CI) using Fisher’s z-transformation and the effective sample size  
- Spearman’s rank correlation (ρ) for robustness to outliers  
- Partial correlation controlling for a linear time trend, to remove potential bias from long-term co-trending  

**Multiple Testing Correction**  
Because the correlation was evaluated across multiple lags, p-values were adjusted using the Holm–Bonferroni method to control the family-wise error rate and ensure statistical rigor across all 25 lag tests.

**Lagged Correlation Analysis**  
To detect delayed responses, the analysis tested lags ranging from −12 to +12 months. Negative lags are included only as a diagnostic reference, to verify that the correlation is directional and not symmetric, confirming that the relationship originates from upstream hydrologic conditions rather than downstream demand.

At each lag, the Pearson, Spearman, and partial correlations were computed, along with their 95% confidence intervals.  
The resulting lag–correlation curve visualizes how the correlation strength evolves with increasing lag and identifies the time offset producing the strongest relationship.

**Visualization**  
Two diagnostic figures were generated:
1. **Lag–correlation plot:** Displays Pearson r versus lag (months) with 95% confidence intervals and highlights statistically significant points (Holm–Bonferroni p ≤ 0.05).  
2. **Time-series plot:** Shows the seasonally adjusted z-scores for Lake Mead elevation and California’s CU over the full 2000–2024 period, allowing visual comparison of their co-fluctuations and long-term trends.

Together, these analyses reveal the characteristic lag and persistence of the upstream–downstream connection between Lake Mead’s storage and California’s consumptive water use.


```python
# --- Imports & Plot Style ---
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import pearsonr, spearmanr, norm, t as tdist

plt.rcParams.update({
    "figure.figsize": (9, 4.6),
    "axes.spines.top": False,
    "axes.spines.right": False,
    "axes.grid": True,
    "grid.linestyle": "--",
    "grid.alpha": 0.3,
    "axes.titlesize": 12,
    "axes.labelsize": 11,
    "xtick.labelsize": 10,
    "ytick.labelsize": 10
})

# Load & Align Datasets 
mead = pd.read_csv("data/LakeMead_Elevation_Monthly_2000_2024.csv", parse_dates=["Date"])
mead = mead[["Date", "Elevation_ft"]].copy()

cu   = pd.read_csv("data/California_WaterUse_2000_2024.csv", parse_dates=["Date"])
cu   = cu[["Date", "ConsumptiveUse_af"]].copy()

# Month-end timestamps for perfect alignment
mead["Date"] = mead["Date"].dt.to_period("M").dt.to_timestamp("M")
cu["Date"]   = cu["Date"].dt.to_period("M").dt.to_timestamp("M")

df = (pd.merge(mead, cu, on="Date", how="inner")
        .sort_values("Date")
        .reset_index(drop=True))

# Study period filter
df = df[(df["Date"] >= "2000-01-01") & (df["Date"] <= "2024-12-31")].copy()

print(f"Monthly records (raw): {len(df)}  |  {df['Date'].min().date()} → {df['Date'].max().date()}")

# Drop months with missing values in either variable
df_clean = df.dropna(subset=["Elevation_ft", "ConsumptiveUse_af"]).copy()
print(f"Records after NaN drop: {len(df_clean)}")

# Water Year helper (Oct–Sep) 
df_clean["WY"] = (df_clean["Date"] + pd.offsets.MonthBegin(3)).dt.year

# Seasonal adjustment: monthly anomalies + z-scores 
for col in ["Elevation_ft", "ConsumptiveUse_af"]:
    mo_grp = df_clean.groupby(df_clean["Date"].dt.month)[col]
    # anomalies against monthly climatology
    df_clean[col + "_anom"] = df_clean[col] - mo_grp.transform("mean")
    # standardize the anomalies
    df_clean[col + "_z"] = (df_clean[col + "_anom"] - df_clean[col + "_anom"].mean()) / \
                           df_clean[col + "_anom"].std(ddof=1)

# Choose the adjusted series for analysis
mead_series = df_clean["Elevation_ft_z"]
cu_series   = df_clean["ConsumptiveUse_af_z"]

# Stats helpers 
def effective_n(x, y):
    """
    Approximate effective sample size for correlation in both series.
    neff = n * (1 - rx*ry) / (1 + rx*ry)
    """
    x = pd.Series(x).dropna()
    y = pd.Series(y).dropna()
    n = min(len(x), len(y))
    if n < 4:
        return max(3, n)
    r1x = x.autocorr(lag=1)
    r1y = y.autocorr(lag=1)
    # Handle edge cases
    if r1x is None or np.isnan(r1x): r1x = 0.0
    if r1y is None or np.isnan(r1y): r1y = 0.0
    denom = (1 + r1x * r1y)
    if np.isclose(denom, 0.0):
        return max(3, n)
    neff = n * (1 - r1x * r1y) / denom
    return float(np.clip(neff, 3, n))

def pearson_with_ci_ac(x, y, alpha=0.05):
    """
    Pearson r with CI and p-value using Fisher z but with an effective n that
    accounts (crudely) for autocorrelation.
    """
    valid = pd.DataFrame({"x": x, "y": y}).dropna()
    n_raw = len(valid)
    if n_raw < 3:
        return np.nan, np.nan, n_raw, (np.nan, np.nan), np.nan

    r, _ = pearsonr(valid["x"], valid["y"])
    n_eff = effective_n(valid["x"], valid["y"])

    # Fisher z CI using n_eff
    r_clip = np.clip(r, -0.999999, 0.999999)
    z = np.arctanh(r_clip)
    se = 1 / np.sqrt(n_eff - 3)
    zcrit = norm.ppf(1 - alpha / 2)
    lo, hi = np.tanh([z - zcrit * se, z + zcrit * se])

    # two-sided p using t with df = n_eff - 2
    tval = r * np.sqrt((n_eff - 2) / (1 - r**2)) if abs(r) < 1 else np.inf
    p = 2 * (1 - tdist.cdf(abs(tval), df=max(1, n_eff - 2)))
    return r, p, n_eff, (lo, hi), n_raw

def holm_bonferroni(pvals):
    """
    Holm–Bonferroni adjustment (no external deps).
    Returns adjusted p-values in original order.
    """
    pvals = np.asarray(pvals, dtype=float)
    m = len(pvals)
    order = np.argsort(pvals)                
    adj = np.empty(m, dtype=float)
    prev = 0.0
    for i, idx in enumerate(order):
        rank = i + 1
        adj_p = (m - i) * pvals[idx]         
        adj[idx] = max(adj_p, prev)
        prev = adj[idx]
    # ensure <= 1
    adj = np.minimum(adj, 1.0)
    return adj

def partial_corr_linear_time(x, y):
    """
    Partial correlation of x and y controlling for a linear time trend.
    Regress x and y separately on [1, t], correlate residuals.
    """
    tmp = pd.DataFrame({"x": x, "y": y}).dropna()
    n = len(tmp)
    if n < 5:
        return np.nan, np.nan, n

    # simple linear trend regressor
    t = np.arange(n)
    X = np.column_stack([np.ones(n), t])

    # OLS for x and y
    bx = np.linalg.lstsq(X, tmp["x"].values, rcond=None)[0]
    by = np.linalg.lstsq(X, tmp["y"].values, rcond=None)[0]
    xres = tmp["x"].values - X @ bx
    yres = tmp["y"].values - X @ by

    r, p = pearsonr(xres, yres)
    return r, p, n

# Lag scan (±12 months) 
max_lag = 12
rows = []

for lag in range(-max_lag, max_lag + 1):
    # shift: positive lag = Mead leads CU; negative = CU leads Mead
    xs = mead_series.shift(lag)
    ys = cu_series

    rP, pP, n_eff, (loP, hiP), n_raw = pearson_with_ci_ac(xs, ys)
    # Spearman (rank-based)
    valid = pd.DataFrame({"x": xs, "y": ys}).dropna()
    if len(valid) >= 3:
        rhoS, pS = spearmanr(valid["x"], valid["y"])
    else:
        rhoS, pS = np.nan, np.nan

    # Partial correlation controlling for linear time trend
    rPart, pPart, nPart = partial_corr_linear_time(xs, ys)

    rows.append({
        "lag_months": lag,
        "pearson_r": rP, "pearson_p": pP, "pearson_lo": loP, "pearson_hi": hiP, "n_eff": n_eff, "n_raw": n_raw,
        "spearman_rho": rhoS, "spearman_p": pS,
        "partial_r_time": rPart, "partial_p_time": pPart, "n_partial": nPart
    })

lagdf = pd.DataFrame(rows).sort_values("lag_months").reset_index(drop=True)

# Multiple-testing adjustment across the scanned lags (for Pearson p-values)
mask = lagdf["pearson_p"].notna()
lagdf.loc[mask, "pearson_p_adj"] = holm_bonferroni(lagdf.loc[mask, "pearson_p"].values)

# Identify strongest effects
bestP_idx = lagdf["pearson_r"].abs().idxmax()
bestS_idx = lagdf["spearman_rho"].abs().idxmax()
bestP = lagdf.loc[bestP_idx]
bestS = lagdf.loc[bestS_idx]

print("\n=== Strongest Pearson (autocorr-aware) ===")
print(f"Lag = {int(bestP['lag_months'])} months | r = {bestP['pearson_r']:.3f} "
      f"(95% CI {bestP['pearson_lo']:.3f}–{bestP['pearson_hi']:.3f}) | "
      f"p = {bestP['pearson_p']:.4f} | p_adj = {bestP['pearson_p_adj']:.4f} | n_eff ≈ {bestP['n_eff']:.1f}")

print("\n=== Strongest Spearman (rank) ===")
print(f"Lag = {int(bestS['lag_months'])} months | ρ = {bestS['spearman_rho']:.3f} | p = {bestS['spearman_p']:.4f}")

# Plot 1: Lagged Pearson with 95% CI and Holm–Bonferroni significance
fig, ax = plt.subplots(figsize=(9.5, 4.8))
x = lagdf["lag_months"]
y = lagdf["pearson_r"]
yerr = np.vstack([y - lagdf["pearson_lo"], lagdf["pearson_hi"] - y])

ax.errorbar(x, y, yerr=yerr, fmt="o-", capsize=3)
ax.axhline(0, ls="--", lw=1, color="gray")
ax.axvline(0, ls=":", lw=1, color="gray")
ax.set_xlabel("Lag (months) — positive: Lake Mead leads CU; negative: CU leads Mead")
ax.set_ylabel("Pearson r (seasonally adjusted)")
ax.set_title("Lagged correlation: Lake Mead Elevation vs California Consumptive Use")

# highlight points significant after Holm–Bonferroni at 0.05
sig = (lagdf["pearson_p_adj"] <= 0.05) & lagdf["pearson_p_adj"].notna()
ax.scatter(x[sig], y[sig], s=70, facecolors="none", edgecolors="black", linewidths=1.5, label="Holm–Bonferroni p≤0.05")
ax.legend(loc="best")
plt.show()

# --- Plot 2: Seasonally-adjusted z-scores over time (sanity check) ---
ax = df_clean.plot(x="Date", y=["Elevation_ft_z", "ConsumptiveUse_af_z"], lw=1.2)
ax.set_title("Seasonally-Adjusted Z-scores\nLake Mead Elevation vs California Consumptive Use")
ax.set_ylabel("Z-score (monthly anomaly standardized)")
ax.legend(["Mead (z)", "CA CU (z)"])
plt.show()

# Inspect the lag table
display(lagdf.round(3))
```

The lag–correlation analysis quantifies how fluctuations in Lake Mead’s elevation propagate through the Lower Colorado River system and influence California’s total consumptive use (CU) between 2000 and 2024.

**Hydrologic and Management Implications**

- Shortage declarations and conservation measures are typically implemented on annual cycles, influencing allocations in the following season.  
- California’s CU is dominated by agriculture, where water ordering, planting, and fallowing decisions follow the crop year rather than the calendar month.  
- A drop in Lake Mead elevation during one water year often triggers reduced CU in the subsequent irrigation season, matching the identified lagged response.

**Summary**

The lag–correlation analysis reveals a strong, statistically robust, and temporally delayed linkage between Lake Mead’s elevation and California’s consumptive use.  
Rather than an immediate or purely seasonal effect, the system exhibits a multi-month to annual response lag consistent with the timing of basin-wide operational and policy processes.  
Lake Mead elevation therefore serves as an early indicator of forthcoming adjustments in California’s water consumption, reflecting the coupled hydrologic and institutional feedbacks that govern the Lower Colorado River Basin.

### 4.2 Response Elasticity Analysis  

This analysis examines the elasticity of California’s consumptive water use in response to changes in Lake Mead elevation, using standardized anomalies to assess how strongly and proportionally water-use behavior adjusts to hydrologic variability.  
Two timescales were evaluated: monthly (to capture short-term, operational adjustments) and annual(to reveal longer-term, system-wide responses).

#### Purpose  
To quantify how sensitive California’s consumptive use is to changes in Lake Mead’s hydrologic condition, and whether this sensitivity differs between seasonal fluctuations and year-to-year trends.

#### Concept of Elasticity  
Response elasticity is represented by the slope of a linear regression between standardized anomalies of Lake Mead elevation and consumptive use:

$$
z_{\text{CU}} = a + b \, z_{\text{Mead}}
$$

where:

- $z_{\text{CU}}$ = standardized anomaly in California’s consumptive use  
- $z_{\text{Mead}}$ = standardized anomaly in Lake Mead elevation  
- $b$ = elasticity coefficient (slope), representing the magnitude of change in water use per unit change in elevation anomaly  

A slope ($b$) near 1 indicates a proportional response (one-for-one change), while smaller slopes indicate a dampened or buffered response.

#### Monthly Elasticity (Seasonal-Adjusted)  
Monthly elasticity was calculated using seasonal z-scores. Each month’s data was standardized relative to its own long-term monthly mean and variability (e.g., all Januaries, all Februaries, etc.).  
This removes recurring seasonal effects and isolates how month-to-month deviations in reservoir elevation relate to deviations in water use under typical operating conditions.  
A simple linear regression between the monthly standardized series estimates the short-term elasticity of response.

#### Annual Elasticity (Interannual)  
To capture broader hydrologic relationships, annual averages of Lake Mead elevation and total annual consumptive use were first computed and then standardized across years using overall z-scores.  
The resulting regression represents interannual elasticity, showing how much California’s total yearly water use tends to increase or decrease in relation to annual-scale departures in reservoir elevation.

#### Interpretation Framework  
Comparing monthly and annual elasticity coefficients provides insight into the temporal dynamics of water-use response:  
- The **monthly elasticity** reflects near-term operational adjustments within irrigation seasons.  
- The **annual elasticity** captures slower, cumulative adjustments linked to longer-term hydrologic and policy conditions.  

Together, these analyses describe how California’s consumptive use scales with changes in Lake Mead’s hydrologic state across short- and long-term time horizons.


```python
# ===============================================================
# Response Elasticity
# (a) Monthly elasticity (seasonal z-scores)
# (b) Annual elasticity (overall z-scores)
# ===============================================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

# ------------------------------
# Load & align monthly data
# ------------------------------
DATA_DIR   = "./data"
FILE_WATER = f"{DATA_DIR}/California_WaterUse_2000_2024.csv"
FILE_MEAD  = f"{DATA_DIR}/LakeMead_Elevation_Monthly_2000_2024.csv"

water = pd.read_csv(FILE_WATER, parse_dates=["Date"])
mead  = pd.read_csv(FILE_MEAD,  parse_dates=["Date"])

to_month = lambda s: pd.to_datetime(s).dt.to_period("M").dt.to_timestamp()
water["Date"] = to_month(water["Date"])
mead["Date"]  = to_month(mead["Date"])

df = (mead.set_index("Date")[["Elevation_ft"]]
          .join(water.set_index("Date")[["ConsumptiveUse_af"]], how="inner")
          .dropna()
          .sort_index())

# ------------------------------
# Monthly z-scores (by calendar month to remove seasonality)
# ------------------------------
def zscore_by_month(series: pd.Series) -> pd.Series:
    g  = series.groupby(series.index.month)
    mu = g.transform("mean")
    sd = g.transform(lambda x: x.std(ddof=1))
    return (series - mu) / sd

df["z_elev_m"] = zscore_by_month(df["Elevation_ft"])
df["z_cu_m"]   = zscore_by_month(df["ConsumptiveUse_af"])

# Monthly regression
res_m = stats.linregress(df["z_elev_m"], df["z_cu_m"])
b_m, a_m, r2_m, p_m = res_m.slope, res_m.intercept, res_m.rvalue**2, res_m.pvalue

# ------------------------------
# Annual aggregation & z-scores (overall)
# ------------------------------
annual_elev = mead.groupby(mead["Date"].dt.year)["Elevation_ft"].mean().rename("Elev_ft")
annual_cu   = water.groupby(water["Date"].dt.year)["ConsumptiveUse_af"].sum().rename("CU_af")

df_ann = pd.concat([annual_elev, annual_cu], axis=1).dropna()
df_ann["z_elev_a"] = (df_ann["Elev_ft"] - df_ann["Elev_ft"].mean()) / df_ann["Elev_ft"].std(ddof=1)
df_ann["z_cu_a"]   = (df_ann["CU_af"]   - df_ann["CU_af"].mean())   / df_ann["CU_af"].std(ddof=1)

# Annual regression
res_a = stats.linregress(df_ann["z_elev_a"], df_ann["z_cu_a"])
b_a, a_a, r2_a, p_a = res_a.slope, res_a.intercept, res_a.rvalue**2, res_a.pvalue

# ------------------------------
# Plot — two panels side-by-side
# ------------------------------
fig, axes = plt.subplots(1, 2, figsize=(13.5, 5))
plt.subplots_adjust(wspace=0.32)

# (a) Monthly
x, y = df["z_elev_m"].values, df["z_cu_m"].values
xx = np.linspace(x.min()-1, x.max()+1, 200)
axes[0].scatter(x, y, alpha=0.7, s=22)
axes[0].plot(xx, xx, "--", color="gray", lw=1)
axes[0].plot(xx, a_m + b_m*xx, color="orange", lw=2)
axes[0].set_title(f"(a) Monthly Elasticity  b={b_m:.2f}, R²={r2_m:.2f}")
axes[0].set_xlabel("Lake Mead Elev. (monthly z)")
axes[0].set_ylabel("Consumptive Use (monthly z)")
axes[0].grid(True)

# (b) Annual
xa, ya = df_ann["z_elev_a"].values, df_ann["z_cu_a"].values
xxa = np.linspace(xa.min()-0.5, xa.max()+0.5, 150)
axes[1].scatter(xa, ya, s=44, color="dodgerblue", edgecolor="k")
axes[1].plot(xxa, xxa, "--", color="gray", lw=1)
axes[1].plot(xxa, a_a + b_a*xxa, color="orange", lw=2)
axes[1].set_title(f"(b) Annual Elasticity  b={b_a:.2f}, R²={r2_a:.2f}")
axes[1].set_xlabel("Lake Mead Elev. (annual z)")
axes[1].set_ylabel("Consumptive Use (annual z)")
axes[1].grid(True)

plt.suptitle("Response Elasticity: CA Consumptive Use vs. Lake Mead Elevation (2000–2024)",
             y=1.05, fontsize=13)
plt.show()

print(f"Monthly  -> slope={b_m:.3f}, R²={r2_m:.3f}, p={p_m:.4f}")
print(f"Annual   -> slope={b_a:.3f}, R²={r2_a:.3f}, p={p_a:.4f}")
```

### 4.3 Policy Milestones and Before–After Water Use Analysis  

This analysis evaluates how **California’s total Colorado River diversions and consumptive use** have changed between 2000 and 2024 in relation to major **federal and basin-level water management policies**.  
The goal is to link observed shifts in annual water use to the implementation of key conservation and allocation frameworks that shaped Lower Basin operations during the 21st century.

#### Purpose  
To assess whether major agreements and management actions—such as the Quantification Settlement Agreement (QSA), the 2007 Interim Guidelines, and the 2019 Drought Contingency Plan—correspond to measurable reductions or behavioral adjustments in California’s overall water use.  
The analysis integrates both time-series visualization and before–after statistical testing to identify the timing and magnitude of responses to policy change.

#### Annual Aggregation  
Monthly records of California’s total diversions and consumptive use were aggregated to annual totals, with only complete calendar years (12 months of data) retained.  
All values were converted from acre-feet (af) to million acre-feet per year (MAF/year) for direct comparability with basin management thresholds and federal accounting reports.

#### Policy Milestones  
The following milestones were analyzed as turning points in basin management and conservation strategy:

- **Quantification Settlement Agreement (QSA, 2003):**  
  Established California’s legal entitlement to **4.4 MAF per year** from the Colorado River, ending decades of overallocation.  
  The QSA formalized long-term transfers from agricultural users (especially the Imperial Irrigation District and Palo Verde Irrigation District) to urban entities such as San Diego and the Metropolitan Water District.  
  It also required efficiency improvements and lining of canals to reduce conveyance losses—initiating a new era of quantified water accounting and internal reallocation within the state.

- **Interim Guidelines (2007):**  
  Adopted by the U.S. Bureau of Reclamation to define shortage-sharing rules among Lower Basin states and coordinate operations of Lake Mead and Lake Powell through the “coordinated reservoir operations” framework.  
  These guidelines established formal trigger elevations at Lake Mead for declaring shortages and introduced the concept of Intentionally Created Surplus (ICS), allowing entities to store conserved water in the reservoir for future use.

- **System Conservation Pilot Program (2014–2017):**  
  A voluntary, incentive-based conservation effort jointly funded by the federal government and Lower Basin states.  
  Under this program, major users—including California’s Imperial Irrigation District—were compensated for conserving water that would remain stored in Lake Mead.  
  The program represented an early test of collaborative, multi-state conservation mechanisms to slow reservoir decline during drought years.

- **Drought Contingency Plan (DCP, 2019):**  
  Expanded on the 2007 Interim Guidelines by strengthening mandatory shortage contributionsand revising ICS accounting.  
  The DCP increased the conservation obligations of all Lower Basin states and established additional Lake Mead elevation tiers with corresponding delivery cuts.  
  For California, it marked the first formal commitment to voluntary reductions under specified hydrologic conditions.

- **Tier 1 Shortage Declaration (2022):**  
  The first official federal shortage declaration for the Lower Colorado River Basin, triggered when Lake Mead fell below 1,075 ft elevation.  
  This milestone activated mandatory delivery reductions for Arizona and Nevada and reinforced conservation expectations across the basin, signaling a new phase of crisis management and adaptation.

Each policy year was marked in the time-series plot with vertical reference lines, while the 2014–2017 System Conservation Pilot was shown as a shaded band to represent its active duration.

#### Trend Visualization  
Annual time series of **diversions** and **consumptive use** were plotted together to visualize how both quantities evolved under successive policy regimes.  
Displaying both on a shared axis (in MAF/year) allows comparison of total withdrawals versus actual depletion, revealing how policy interventions and hydrologic stress influenced overall water-use efficiency and behavior.

#### Before–After Statistical Comparison  
For each policy milestone, a **before–after analysis** was performed to quantify the average change in annual diversions and consumptive use.  
The analysis calculated:  

- **Mean before** and **mean after** each policy’s implementation year  
- **Absolute change (ΔMean, MAF)** and **percent change (ΔMean %)**  
- **Mann–Whitney U test p-value** to assess whether the distributions before and after each policy differ significantly  

This nonparametric test evaluates whether observed changes reflect meaningful shifts in central tendency rather than random interannual variability.

#### Interpretation Framework  
By combining visual trends with statistical comparisons, this analysis provides an integrated view of how major water management policies influenced California’s use of Colorado River water.  
It helps identify whether policy interventions coincided with measurable declines in total diversions, reductions in consumptive use, or changes in the variability of annual water use.  
Together, these results contribute to understanding the **policy effectiveness**, **behavioral adaptation**, and **resilience** of California’s water management system under evolving basin-wide constraints.


```python
# %% California annual Diversions & Consumptive Use with policy milestones (y-axis from 0)

import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
from matplotlib.lines import Line2D

# ------------------ 1. Load and aggregate annual data ------------------
# Reads the California monthly water use dataset and computes annual totals
# for both consumptive use and total diversions. Only full years (12 months of data)
# are included to ensure consistency across the record.

use = pd.read_csv("data/California_WaterUse_2000_2024.csv", parse_dates=["Date"])[
    ["Date", "ConsumptiveUse_af", "Diversion_af"]
]
df = use.copy()
df["Year"] = df["Date"].dt.year

# Compute annual totals and keep only complete years
annual = (
    df.groupby("Year")
      .agg(n_months=("ConsumptiveUse_af", "count"),
           CU_total_af=("ConsumptiveUse_af", "sum"),
           Div_total_af=("Diversion_af", "sum"))
      .reset_index()
)
annual = annual[annual["n_months"] == 12].copy()

# Convert acre-feet to million acre-feet (MAF)
annual["CU_MAF"] = annual["CU_total_af"] / 1e6
annual["Div_MAF"] = annual["Div_total_af"] / 1e6

# ------------------ 2. Define key policy milestones ------------------
# Each policy event is represented by a vertical line with a distinct color and style.
# The shaded region (2014–2017) indicates the period of the System Conservation Pilot.

policy_years = {
    "QSA (4.4 MAF Cap)":         {"year": 2003, "color": "darkred",  "ls": "-."},
    "Interim Guidelines (2007)": {"year": 2007, "color": "maroon",   "ls": "-"},
    "Drought Contingency Plan":  {"year": 2019, "color": "firebrick","ls": ":"},
    "Tier 1 Shortage":           {"year": 2022, "color": "crimson",  "ls": "--"},
}
pilot_start, pilot_end = 2014, 2017  # System Conservation Pilot duration

# ------------------ 3. Plot annual Diversions and Consumptive Use ------------------
fig, ax = plt.subplots(figsize=(11, 5.5))

# Plot annual consumptive use (blue) and diversions (green)
ax.plot(
    annual["Year"], annual["CU_MAF"],
    marker="o", lw=2, color="steelblue",
    label="Consumptive Use (MAF/year)"
)
ax.plot(
    annual["Year"], annual["Div_MAF"],
    marker="s", lw=2, color="seagreen",
    label="Diversions (MAF/year)"
)

# Highlight the 2014–2017 Pilot Program as a shaded band
ax.axvspan(pilot_start, pilot_end + 1, color="lightcoral", alpha=0.15)

# Add vertical lines for major policy milestones
for label, spec in policy_years.items():
    ax.axvline(spec["year"], color=spec["color"], ls=spec["ls"], lw=1.8)

# ------------------ 4. Format axes and labels ------------------
ax.set_title("California Annual Diversions and Consumptive Use with Major Policy Milestones", fontsize=12, pad=12)
ax.set_xlabel("Year", fontsize=10)
ax.set_ylabel("Million Acre-Feet per Year", fontsize=10)
ax.grid(True, alpha=0.3)

# Display all years on the x-axis
ax.set_xlim(annual["Year"].min() - 0.5, annual["Year"].max() + 0.5)
ax.set_xticks(annual["Year"])
plt.xticks(rotation=45, ha="right")

# Start the y-axis at zero to better show magnitude of change
ax.set_ylim(bottom=0)

# ------------------ 5. Create a combined legend ------------------
# The legend includes both water-use series, the shaded Pilot period,
# and each policy event with its corresponding style.

legend_items = [
    Line2D([0], [0], color="steelblue", lw=2, marker="o", label="Consumptive Use"),
    Line2D([0], [0], color="seagreen", lw=2, marker="s", label="Diversions"),
    Rectangle((0,0), 1, 1, facecolor="lightcoral", alpha=0.15, label="System Conservation Pilot (2014–2017)"),
]
for label, spec in policy_years.items():
    legend_items.append(Line2D([0], [0], color=spec["color"], ls=spec["ls"], lw=2, label=label))

# Place legend outside the right margin to prevent overlap with the plot area
ax.legend(
    handles=legend_items,
    frameon=False, fontsize=9, ncol=1,
    loc="center left", bbox_to_anchor=(1.02, 0.5)
)

plt.tight_layout()
plt.show()


# %% Before–After summary table (Consumptive Use & Diversions) — essential statistics only

import pandas as pd
from scipy.stats import mannwhitneyu

# ------------------ 6. Prepare annual dataset ------------------
# Ensures that the annual summary table is available. 
# If not already defined in memory, it rebuilds the 'annual' DataFrame.

try:
    _ = annual[["Year", "CU_MAF", "Div_MAF"]]
except Exception:
    use = pd.read_csv("data/California_WaterUse_2000_2024.csv", parse_dates=["Date"])[
        ["Date", "ConsumptiveUse_af", "Diversion_af"]
    ]
    use["Year"] = use["Date"].dt.year
    annual = (use.groupby("Year")
                 .agg(n_months=("ConsumptiveUse_af","count"),
                      CU_total_af=("ConsumptiveUse_af","sum"),
                      Div_total_af=("Diversion_af","sum"))
                 .reset_index())
    annual = annual[annual["n_months"] == 12].copy()
    annual["CU_MAF"]  = annual["CU_total_af"]  / 1e6
    annual["Div_MAF"] = annual["Div_total_af"] / 1e6

# ------------------ 7. Define key policy years ------------------
# Each policy year marks the start of a management or conservation regime.
# The before/after comparison quantifies the average change in water use
# following each policy implementation.

policies = {
    "QSA (2003)": 2003,
    "Interim Guidelines (2007)": 2007,
    "System Conservation Pilot (2014)": 2014,
    "Drought Contingency Plan (2019)": 2019,
    "Tier 1 Shortage (2022)": 2022,
}

# ------------------ 8. Define helper function for before–after comparison ------------------
# For each policy, calculates the mean consumptive use or diversion
# before and after the policy year. Also computes the absolute (ΔMean)
# and percent change (ΔMean %) and performs a Mann–Whitney U test
# to test whether the two distributions differ significantly.

def before_after_table(df, value_col, metric_label):
    rows = []
    for name, year in policies.items():
        pre  = df.loc[df["Year"] < year,  value_col].dropna()
        post = df.loc[df["Year"] >= year, value_col].dropna()
        if len(pre) < 3 or len(post) < 3:
            continue  # skip policies with insufficient data
        mean_pre, mean_post = pre.mean(), post.mean()
        d_maf  = mean_post - mean_pre
        d_pct  = (mean_post / mean_pre - 1) * 100
        U, p   = mannwhitneyu(pre, post, alternative="two-sided")
        rows.append({
            "Policy": name,
            "Year": year,
            "Metric": metric_label,
            "Mean Before (MAF)": round(mean_pre, 2),
            "Mean After (MAF)":  round(mean_post, 2),
            "ΔMean (MAF)":       round(d_maf, 2),
            "ΔMean (%)":         round(d_pct, 1),
            "p-value":           round(p, 3),
        })
    return pd.DataFrame(rows)

# ------------------ 9. Apply the comparison for both metrics ------------------
tbl_cu  = before_after_table(annual, "CU_MAF",  "Consumptive Use")
tbl_div = before_after_table(annual, "Div_MAF", "Diversions")

# Combine results into one summary table and sort chronologically
policy_changes_simple = (pd.concat([tbl_cu, tbl_div], ignore_index=True)
                           .sort_values(["Year", "Metric"])
                           .reset_index(drop=True))

# ------------------ 10. Display final summary table ------------------
# The resulting table summarizes the change in mean consumptive use and diversions
# before and after each major policy milestone.

display(policy_changes_simple)



```

**Discussion of Results: Policy Impacts on California’s Colorado River Water Use (2000–2024)**

The combined visualization and before–after statistics reveal a clear sequence of policy-driven reductions in both **Diversions** and **Consumptive Use** since the early 2000s.  
Overall, California’s total diversions declined from an average of **~5.8 MAF before 2003** to **~4.5–4.8 MAF afterward**, while net consumptive use fell from **~5.3 MAF** to **~4.0–4.3 MAF**.



**1. Quantification Settlement Agreement (QSA, 2003)**
The QSA represented the single largest structural reduction in California’s water use.  
Immediately after its implementation, both diversions (**–16.7%**) and consumptive use (**–18.3%**) showed highly significant decreases (p = 0.001).  
This sharp drop marks the enforcement of the 4.4 MAF cap, as agricultural transfers and conservation programs in the Imperial and Coachella valleys began to take effect.



**2. Interim Guidelines (2007)**
Following the 2007 Guidelines, California’s mean diversions and consumptive use declined by about **10%**, though only the change in **diversions** was statistically significant (p = 0.025).  
This suggests that while coordinated reservoir operations limited deliveries to maintain Lake Mead stability, on-the-ground efficiency or behavioral changes in net water use lagged behind structural allocation shifts.



**3. System Conservation Pilot Program (2014–2017)**
The Pilot Program period corresponds to moderate but notable additional reductions in both metrics, with **–9.4%** for diversions (p = 0.009) and **–7.9%** for consumptive use (p = 0.085).  
Although the consumptive-use change was only marginally significant, these results indicate meaningful voluntary savings achieved through compensated conservation projects and temporary fallowing agreements.


 **4. Drought Contingency Plan (2019)**
Post-2019, both diversions and consumptive use dropped by roughly **10%**, with statistically significant changes (p = 0.004–0.030).  
This reflects the early effectiveness of the DCP’s voluntary conservation commitments and expanded Intentionally Created Surplus (ICS) participation, even before mandatory cuts applied to California.


 **5. Tier 1 Shortage (2022)**
The 2022 shortage declaration marked the first formal shortage condition in the Lower Basin.  
California’s changes after this point were smaller and not statistically strong for consumptive use (p = 0.177), though diversions continued to decline slightly (p = 0.046).  
Because California’s reductions remain voluntary under Tier 1, the post-2022 decline likely reflects **continued cooperative conservation** rather than enforced curtailment.

 **Summary Insight**
Across all policies, California has achieved sustained reductions in both **diversions and consumptive use** since 2003, demonstrating long-term adaptation to basin-wide shortage conditions.  
The most significant impacts align with **structural and incentivized programs** (QSA and SCPP), while later voluntary conservation efforts (DCP and Tier 1) continued the downward trend but with smaller magnitudes and higher variability.

Together, the analysis illustrates how **policy evolution—from mandatory caps to voluntary conservation—progressively reduced water use intensity** while maintaining agricultural and urban reliability within California’s 4.4 MAF entitlement.
